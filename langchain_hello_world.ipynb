```python
from IPython import get_ipython
from IPython.display import display
# %%
from google.colab import userdata
GEMINI_API_KEY = userdata.get('GOOGLE_API_KEY')

!pip install langchain-google-genai

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate

# Define the prompts
first_prompt = PromptTemplate(
    input_variables = ["question"],
    template = "Write 5 key points about of this {question}"
)

second_prompt = PromptTemplate(
    input_variables = ["first_response"],
    template = "explain each points in detail of this {first_response}"
)

# Initialize the language model
llm = ChatGoogleGenerativeAI(
    api_key = GEMINI_API_KEY,
    model = 'gemini-1.5-flash',
    temperature = 0.7
)

# Create the chains (This cell must be run before using the chains)
first_chain = first_prompt | llm
second_chain = second_prompt | llm

#Invoke the first chain
first_response = first_chain.invoke({"question": "What is ML"})
print("First chain", first_response.content)

#Invoke the second chain
second_response = second_chain.invoke({"first_response" : first_response.content})
print("Second chain", second_response.content)
# %%
from google.colab import userdata
GEMINI_API_KEY = userdata.get('GOOGLE_API_KEY')
# %%
!pip install langchain_google_genai
# %%
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
# %%
# Define image generation prompts
first_prompt = PromptTemplate(
    input_variables=["question"],  # Corrected typo: input_varisblr to input_variables
    template="Change this {question} into a good image generate image prompt ?"
)

second_prompt = PromptTemplate(
    input_variables=['first_response'],  # Corrected typo: input_variable to input_variables
    template="Read this {first_response} and summarize into two line prompt for image generation "
)

third_prompt = PromptTemplate(
    input_variables=['second_response'],  # Corrected typo: input_variable to input_variables
    template="Read this {second_response} and summarize into two line prompt for image generation "
)

# %%
llm = ChatGoogleGenerativeAI(
    api_key=GEMINI_API_KEY,
    model="gemini-1.5-flash",
    temperature=0.5
)
# %%
# Create image generation chains (This cell must be run before using the chains)
first_chain = first_prompt | llm
second_chain = second_prompt | llm
third_chain = third_prompt | llm
# %%
!pip install diffusers
!pip install torch torchvision
!pip install transformers
!pip install accelerate
# %%
from diffusers import StableDiffusionPipeline
import torch

# Invoke the chains and generate images
first_response = first_chain.invoke({"question": input("Write a image generate prompt ")})
print("First chain", first_response.content)

second_response = second_chain.invoke({"first_response": first_response.content})
print("Second_response", second_response.content)

# Load the Stable Diffusion pipeline
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
pipe.to("cuda")  # Use CUDA for faster processing

# Generate an image from text
image = pipe(second_response.content).images[0]

# Save the image
image.save("generated_image.png")
```
